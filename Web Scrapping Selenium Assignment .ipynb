### Q2] Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data.

```python
#Importing libraries
from selenium.common.exceptions import NoSuchElementException
import numpy as np
```

```python
#Connecting to the web driver
driver1=webdriver.Chrome(r"C:\Users\MUNGI\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
#Connecting to the website with the help of url
url="https://www.naukri.com/"
driver1.get(url)
```

```python
#Finding element for job search bar 
search_job=driver1.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys("Data Scientist")
```

```python
#finding element for job location bar
search_loc=driver1.find_element_by_id('qsb-location-sugg')
search_loc.send_keys("Bangalore")
```

```python
#do click using xpath function
search_btn=driver1.find_element_by_xpath("//button[@class='btn']")
search_btn.click()
```

```python
#Fetching urls
urls=[]
for i in driver1.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute("href"))
```

```python
#Displaying all urls
urls
```

```python
#Defining variables
job_title=[]
job_location=[]
company_name=[]
job_description=[]
```

```python
#Fetching data from the urls
for i in urls[:10]:
    driver1.get(i)
    time.sleep(3)
    #Fetching job title
    try:
        job=driver1.find_element_by_xpath("//h1[@class='jd-header-title']")
        job_title.append(job.text)
    except NoSuchElementException as e:
        job_title.append('-')
        
    #Fetching job location
    try:
        job_loc=driver1.find_element_by_xpath("//div[@class='loc']")
        job_location.append(job_loc.text)
    except NoSuchElementException as e:
        job_location.append('-')
        
    #Fetching company name
    try:
        job_com=driver1.find_element_by_xpath("//div[@class='jd-header-comp-name']")
        company_name.append(job_com.text)
    except NoSuchElementException as e:
        company_name.append('-')
        
     #Fetching job description
    try:
        job_desc=driver1.find_element_by_xpath("//section[@class='job-desc']")
        job_description.append(job_desc.text)
    except NoSuchElementException as e:
        job_description.append('-')
```

```python
#Printing length of each variable
print(len(job_title),len(job_location),len(company_name),len(job_description))
```

```python
#Displaying job title
job_title
```

```python
#Displaying job location
job_location
```

```python
#Displaying company name
company_name
```

```python
#Displaying job description
job_description
```

```python
#Creating dataframe for scrapped data
job=pd.DataFrame({})
job['Title']=job_title
job['Location']=job_location
job['Company Name']=company_name
job['Description']=job_description
```

```python
job
```
