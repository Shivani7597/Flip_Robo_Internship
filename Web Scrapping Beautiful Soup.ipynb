### Question 1] Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
#Sending request to the webpage server to get the source code of the page which is to be scrapped by using the request library
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

Having a look at the webpage source code downloaded above 

```python
page.content
```

Parsing this source code by using BeautifulSoup and the parser used here is html.parser as the source code is in html.

```python
soup=BeautifulSoup(page.content)
soup
```

```python
#getting indentation to the code
print(soup.prettify())
```

```python
#Extracting the header tags from the web page
header=soup.find_all('span',class_="mw-headline")
header
```

```python
#Displaying all header text
header_tags=[]

for i in header:
    header_tags.append(i.get_text().replace('\xa0',''))
    
header_tags
```

```python
len(header_tags)
```

```python
import pandas as pd
header=pd.DataFrame({})
header['Header Tags of Wikipedia']=header_tags
```

```python
header
```

### Question 4] Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
#Sending request to the webpage server to get the source code of the page which is to be scrapped by using the request library
page2=requests.get("https://bookpage.com/reviews")
page2
```

```python
#Importing library
soup = BeautifulSoup(page2.text, 'html.parser')
```

```python
#Extracting book names from the web page
book=soup.find_all('h4',class_="italic")
book
```

```python
#Displaying all movie names
book_names=[]

for i in book:
    book_names.append(i.get_text().replace('\n',''))
    
book_names
```

```python
#Extracting reviews of books from the web page
review=soup.find_all('p',class_="excerpt")
review
```

```python
#Displaying all reviews
reviews=[]

for i in review:
    reviews.append(i.get_text().replace('\n',''))
    
reviews
```

```python
#print length of elements present in each attribute
print(len(book_names),len(author_names),len(genres),len(reviews))
```

```python
#creating data frame for books
import pandas as pd
books=pd.DataFrame({})
books['Book Name']=book_names
books['Author']=author_names
books['Genre']=genres
books['Review']=reviews
```

```python
#Selecting any 5 books
books.drop([1,5,3,2,9],axis=0,inplace=True)
```

```python
#Details of any 5 books
books
```
