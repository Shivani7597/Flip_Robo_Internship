### 2] In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#Creating empty list to fill info
brand_name1=[]
name_of_product1=[]
rating1=[]
no_ratings1=[]
price1=[]
return_exchange1=[]
expected_delivery1=[]
availability1=[]
other_details1=[]
urls1=[]
```

```python
#Fetching urls of first three pages
for i in range(0,3):
    for j in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        urls1.append(j.get_attribute("href"))
        driver.find_element_by_xpath('//li[@class="a-last"]')
        time.sleep(5)
```

```python
urls1
```

```python
#Fetching data from the urls
for i in urls1:
    driver.get(i)
    time.sleep(5)
    #Fetching brand name
    try:
        brand_name=driver.find_element_by_xpath("//td[@class='a-span9']")
        brand_name1.append(brand_name.text)
    except NoSuchElementException as e:
        brand_name1.append('-')
        
    #Fetching Name of the Product
    try:
        name_of_product=driver.find_element_by_id("productTitle")
        name_of_product1.append(name_of_product.text)
    except NoSuchElementException as e:
        name_of_product1.append('-')
        
    #Fetching Rating
    try:
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        rating1.append(rating.text)
    except NoSuchElementException as e:
        rating1.append('-')
        
    #Fetching No. of Ratings
    try:
        no_ratings=driver.find_element_by_id("acrCustomerReviewText")
        no_ratings1.append(no_ratings.text)
    except NoSuchElementException as e:
        no_ratings1.append('-')
        
    #Fetching Price
    try:
        price=driver.find_element_by_xpath("//td[@class='a-span12']")
        price1.append(price.text)
    except NoSuchElementException as e:
        price1.append('-')
        
    #Fetching Return/Excahnge
    try:
        return_exchange=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none icon-content']")
        return_exchange1.append(return_exchange.text)
    except NoSuchElementException as e:
        return_exchange1.append('-')
        
    #Fetching Expected Delivery
    try:
        expected_delivery=driver.find_element_by_id("dynamicDeliveryMessage_feature_div")
        expected_delivery1.append(expected_delivery.text)
    except NoSuchElementException as e:
        expected_delivery1.append('-')
        
    #Fetching Availability
    try:
        availability=driver.find_element_by_id("availability")
        availability1.append(availability.text.replace("\n"," "))
    except NoSuchElementException as e:
        availability1.append('-')
        
    #Fetching Other Details
    try:
        other_details=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']")
        other_details1.append(other_details.text)
    except NoSuchElementException as e:
        other_details1.append('-')
```

```python
#Checking length of records in each variable
print(len(brand_name1),len(name_of_product1),len(rating1),len(no_ratings1),len(price1),len(return_exchange1),len(expected_delivery1),len(availability1),len(other_details1),len(urls1))
```

```python
#Creating dataframe
guitar=pd.DataFrame({})
guitar['Brand Name']=brand_name1
guitar['Name of Product']=name_of_product1
guitar['Rating']=rating1
guitar['No. of Ratings']=no_ratings1
guitar['Price']=price1
guitar['Return/ Exchange']=return_exchange1
guitar['Expected Delivery']=expected_delivery1
guitar['Availbility']=availability1
guitar['Other Details']=other_details1
guitar['Product URL']=urls1
```
