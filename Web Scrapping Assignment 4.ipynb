### 1] Scrape the details of most viewed videos on YouTube from Wikipedia:
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
 
You need to find following details:

A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\MUNGI\Downloads\chromedriver_win32\chromedriver.exe")
#Opening webpage using given url
driver.get(' https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#To view most viewed videos on YouTube, clicking on "Top Videos" in the Contents section
topv=driver.find_element_by_xpath('//li[@class="toclevel-1 tocsection-2"]/a')
topv.click()
```

```python
#Defining Variable list to store values
rank=[]
name=[]
artist=[]
date=[]
views=[]
```

```python
#Fetching Rank of the video
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")[:30]:
        rank.append(i.text)
except NoSuchElementException as e:
    rank.append('-')
        
#Fetching video name
try:
    for j in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")[:30]:
        name.append(j.text)
except NoSuchElementException as e:
    name.append('-')
        
#Fetching name of artist
try:
    for k in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")[:30]:
        artist.append(k.text)
except NoSuchElementException as e:
    artist.append('-')
        
#Fetching date of video uploaded 
try:
    for l in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")[:30]:
        date.append(l.text)
except NoSuchElementException as e:
    date.append('-')
        
#Fetching no. of views on video
try:
    for m in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")[:30]:
        views.append(m.text)
except NoSuchElementException as e:
    views.append('-')
```

```python
#Closing driver
driver.close()
```

```python
#Checking length of each variable
print(len(rank),len(name),len(artist),len(date),len(views))
```

```python
#Creating dataframe1
top_videos=pd.DataFrame({})
top_videos['Rank']=rank
top_videos['Video Name']=name
top_videos['Artist Name']=artist
top_videos['Upload Date']=date
top_videos['Views(billions)']=views

#Setting "Rank" as index
top_videos.set_index("Rank",inplace=True)
```

```python
#Displaying top most viewed videos on youtube
top_videos
```

D) Language used

```python
#Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\MUNGI\Downloads\chromedriver_win32\chromedriver.exe")
#Opening home page of github
driver.get(' https://github.com/')
```

```python
#From home page, using the explore drop down list, clicking on the trendin option

#First selecting explore to visualize drop down list
explore=driver.find_element_by_xpath('//ul[@class="d-lg-flex list-style-none"]/li[4]')
explore.click()

#Selecting Trending from drop down list
trending=driver.find_element_by_xpath('//ul[@class="list-style-none mb-3"][2]/li[3]/a')
trending.click()
```

```python
#Fetching urls
urls=[]
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    urls.append(i.get_attribute("href"))
```

```python
urls
```

```python
#Defining empty lists to store values
title=[]
desc=[]
contri_count=[]
language=[]
```

```python
#Fetching data from the urls
for i in urls:
    driver.get(i)
    time.sleep(5)
    
    #Fetching Repository title
    try:
        ti=driver.find_element_by_xpath("//h1[@class=' d-flex flex-wrap flex-items-center break-word f3 text-normal']")
        title.append(ti.text.replace("\n"," "))
    except NoSuchElementException as e:
        title.append('-')
        
    #Fetching Repository description
    try:
        de=driver.find_element_by_xpath("//p[@class='f4 mt-3']")
        desc.append(de.text)
    except NoSuchElementException as e:
        desc.append('-')
        
    #Fetching Contributors count
    try:
        count=driver.find_element_by_xpath("//div[@class='BorderGrid-row'][3]/div/h2/a/span")
        contri_count.append(count.text)
    except NoSuchElementException as e:
        contri_count.append('-')
        
    #Fetching Language used
    try:
        lang=driver.find_element_by_xpath("//div[@class='BorderGrid-row'][4]/div/ul")
        language.append(lang.text.replace("\n"," "))
    except NoSuchElementException as e:
        language.append('-')
        
    
```

```python
#Checking records in each variable
print(len(title),len(desc),len(contri_count),len(language))
```

```python
#Creating dataframe
github=pd.DataFrame({})
github['Repository Title']=title
github['Repository Description']=desc
github['Contributors Count']=contri_count
github['Language Used']=language
```

```python
#Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\MUNGI\Downloads\chromedriver_win32\chromedriver.exe") 
time.sleep(2)

#Opening the naukri.com
url = "https://www.naukri.com/"
driver.get(url)
time.sleep(2)

#Clicking on recruiters link
recruiters=driver.find_element_by_xpath("//a[@title='Search Recruiters']")
driver.get(recruiters.get_attribute('href'))
time.sleep(1)

#Accesing search bar
search=driver.find_element_by_xpath("//input[@class='sugInp']")
search.send_keys("Data Science")
time.sleep(1)

#Clicking on search button
button=driver.find_element_by_xpath("//button[@id='qsbFormBtn']")
button.click()
time.sleep(1)
```
